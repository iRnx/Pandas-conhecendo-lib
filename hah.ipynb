```python
import pandas as pd
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados_df = pd.read_csv(url, sep=';')
display(dados_df)
```

```python
query = dados_df.query('Tipo == "Apartamento"')

query = query[['Quartos']].mean()
display(query)
```

```python
dados_df.Bairro.nunique()
```

```python
dados_df.groupby('Tipo')['Valor'].mean().to_frame().sort_values(by='Tipo', ascending=True)
```

```python
v = dados_df.groupby('Bairro')['Valor'].mean().to_frame().sort_values(by='Valor', ascending=False).head()
display(v)
v.plot(kind='barh', figsize=(14, 10), color='orange')
```
