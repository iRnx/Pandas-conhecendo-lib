Removendo os imóveis comerciais

```python
dados_df.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                    'Prédio Inteiro', 'Loja/Salão',
                    'Galpão/Depósito/Armazém', 'Casa Comercial',
                    'Terreno Padrão', 'Loja Shopping/ Ct Comercial', 'Box/Garagem',
                    'Chácara', 'Loteamento/Condomínio', 'Sítio', 'Pousada/Chalé', 'Hotel',
                    'Indústria']

# como essa query esta trazendo apenas os dados dos imoveis comerciais
dados_df.query('@imoveis_comerciais in Tipo')
```

```python
# Para não trazer estes dados basta arrumar a logica dentro da query
tipos_residenciais = dados_df.query('@imoveis_comerciais not in Tipo')
display(tipos_residenciais.head())
```

```python
tipos_residenciais.Tipo.unique()
```

```python
tipo_valor_df = tipos_residenciais.groupby('Tipo')[['Valor']].mean().sort_values(by='Valor')

tipo_valor_df.plot(kind='barh', figsize=(15, 10), color='purple')
```

Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
tipos_residenciais.Tipo.unique()
```

```python
tipos_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values(by='proportion')
```

```python
percentual_tipo = tipos_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')
percentual_tipo.plot(kind='bar', figsize=(14, 10), color='orange', xlabel='Tipos', ylabel='Percentual')
```

Selecionando apenas os imoveis do tipo apartamnto

```python
dados_df = tipos_residenciais.query('Tipo == "Apartamento"')
display(dados_df)
```

```python
# Calcular a média de quartos por apartamento;
dados_df['Quartos'].mean()
```

```python
# Conferir quantos bairros únicos existem na nossa base de dados;
len(dados_df.Bairro.unique())
dados_df['Bairro'].nunique()
```

```python
# Analisar quais bairros possuem a média de valor de aluguel mais elevadas;
dados_df.groupby('Bairro')['Valor'].mean().sort_values(ascending=False)
```

```python
# Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevadas.
df_bairros = dados_df.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False).head()
df_bairros.plot(kind='barh', figsize=(14,10), color='blue')

```

Lidando com dados Nulos

```python
# Verificando se tem dados nulos, True significa que tem dados nulos
dados_df.isnull()
```

```python
# Verificar em quais colunas tem dados nulos.
dados_df.isnull().sum()
```

Removendo registros

```python
# Fazendo uma query para trazer dados onde o valor e condominio sejam 0 #
dados_df.query('Valor == 0 | Condominio == 0')
```

```python
# Pegamos os indices de cada linha q temos que remover
registros_a_remover = dados_df.query('Valor == 0 | Condominio == 0').index
display(registros_a_remover)
```

```python
# axis 0 para informar que queremos remover linhas, linhas=0, colunas=1
dados_df.drop(registros_a_remover, axis=0, inplace=True)
```
