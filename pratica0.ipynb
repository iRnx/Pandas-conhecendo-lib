```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
aluguel_df = pd.read_csv(url, sep=';')
display(aluguel_df)
```

```python
grafico_aluguel_df = aluguel_df.groupby('Tipo')[['Valor']].mean().sort_values(by='Valor', ascending=True)
grafico_aluguel_df.plot(kind='bar', figsize=(14, 10), color='orange', ylabel='Valor', xlabel='Imóveis')
```

```python
aluguel_df.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                    'Prédio Inteiro', 'Loja/Salão',
                    'Galpão/Depósito/Armazém', 'Casa Comercial',
                    'Terreno Padrão', 'Loja Shopping/ Ct Comercial', 'Box/Garagem',
                    'Chácara', 'Loteamento/Condomínio', 'Sítio', 'Pousada/Chalé', 'Hotel',
                    'Indústria']

imoveis_residenciais = aluguel_df.query('@imoveis_comerciais not in Tipo')
display(imoveis_residenciais)
```

```python
valor = imoveis_residenciais.groupby('Tipo')['Valor'].mean().to_frame().sort_values(by='Valor', ascending=True)
display(valor)
valor.plot(kind='bar', figsize=(14, 10), color='black')
```

```python
grafico = imoveis_residenciais.Tipo.value_counts().to_frame().sort_values(by='count', ascending=True)
display(grafico)
grafico.plot(kind='bar', figsize=(14, 10), color='purple')
```

```python
apartamentos = imoveis_residenciais.query('Tipo == "Apartamento" | Tipo == "Casa de Condomínio"')
display(apartamentos)
```
